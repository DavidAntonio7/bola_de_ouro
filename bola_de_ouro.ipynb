##quantificando o valor das posicoes no rank

```python
df[df.Year == 2002].head()
```

```python
df[df.Year == 2006]
```

```python
listaNomes = []
for i in df.Player:
  if i in listaNomes:
    continue
  else:
    listaNomes.append(i)
```

```python
len(listaNomes)
```

```python
df.shape[0]
```

```python
df[df.Rank == 1]
```

```python
df.Player.iloc[1]
```

```python
percentualTotal = []
for jogador in listaNomes: 
  percentualTotal.append(df[df.Player.str.contains(jogador)].Percent.sum())
```

```python
len(percentualTotal)
```

```python
participacao = []
for jogador in listaNomes: 
  participacao.append(df[df.Player.str.contains(jogador)])
```

```python
participacao[1]
```

```python
len(participacao )
```

```python
participacao[1].Rank.values[1]
```

```python
participacao[1].Rank.values
```

```python
somaRank = []
for jogador in range(0,len(participacao)):
  somaRank.append(0)
  for i in participacao[jogador].Rank.values:
    if  i <= 30:
      somaRank[jogador] += 31 - i
    else:
      somaRank[jogador] += 1
```

```python
len(somaRank)
```

```python
somaRank[1]
```

```python
resultado = pd.DataFrame({'soma_Percentuais' : percentualTotal, 'calculo_Rank' : somaRank} , index = listaNomes)
```

```python
resultado.head()
```

```python
resultado.head().index
```
